```python
# !pip install seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from database_management import Offer, Token, Token_Value_Usd
```

```python
x,y,hue=zip(*Offer.query.with_entities(Offer.datetime, Offer.price, Offer.token_id).all())
x=np.array(x)
y=np.array(y)
hue=np.array(hue)
```

```python
fig, ax= plt.subplots(3,2,figsize=(30,20))
ax=ax.flatten()
for i, n in enumerate(np.unique(hue)): 
    sns.scatterplot(x=x[hue==n], y=y[hue==n],ax=ax[i-1])
```
